```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
list(df.columns.values)
```

```python
df.dtypes
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')
```

```python
df.isnull().sum()
```

```python
#Replace null values in Total Charges with medium values of the column

df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].mean())
```

```python
df["TotalCharges"].isnull().sum()
```

```python
#Delete the customers' id columns
del df['customerID']
```

```python
df.head()
```

```python
target_col = ["Churn"]

cat_cols   = df.nunique()[df.nunique() < 10].keys().tolist()

cat_cols   = [x for x in cat_cols if x not in target_col]

binary_cols   = df.nunique()[df.nunique() == 2].keys().tolist()

multi_cols = [i for i in cat_cols if i not in binary_cols]
```

```python
binary_cols
```

```python
multi_cols
```

```python
# from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

#Encode Yes/No columns
for cols in binary_cols :
    df[cols] = le.fit_transform(df[cols])
    
#Encode multivalues columns
df = pd.get_dummies(data = df,columns = multi_cols )
```

```python
df.head()
```

```python
#Standardize numeric columns
from sklearn.preprocessing import StandardScaler

num_cols   = [x for x in df.columns if x not in cat_cols + target_col]

ss = StandardScaler()

scl = ss.fit_transform(df[num_cols])
scl = pd.DataFrame(scl, columns=num_cols)

#Drop current numeric columns
df = df.drop(columns = num_cols,axis = 1)

df = df.merge(scl, left_index=True, right_index=True, how = "left")
```

```python
df.head()
```

# Modeling

```python
from sklearn.model_selection import train_test_split

X_train = df.drop(['Churn'], axis=1)
y_train = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=.2)
```

```python
# Support vector machine

from sklearn.metrics import classification_report
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
pred_svc = svc.predict(X_test)

print(classification_report(y_test,pred_svc))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 200)
rfc.fit(X_train,y_train)
pred_rfc = rfc.predict(X_test)

print(classification_report(y_test,pred_rfc))
```

```python
from sklearn.metrics import confusion_matrix
#Confusion matrix for the random forest classification
print(confusion_matrix(y_test, pred_rfc))
```

```python
#Using GridSearchCV to fine-tune the above model

from sklearn.model_selection import GridSearchCV
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc, param_grid=param, scoring='accuracy', cv=10)

grid_svc.fit(X_train, y_train)
```

```python
#Best parameters for our svc model
grid_svc.best_params_
```

```python
#Run our SVC again with the best parameters.
svc2 = SVC(C = 0.9, gamma =  0.1, kernel= 'linear')
svc2.fit(X_train, y_train)
pred_svc2 = svc2.predict(X_test)
print(classification_report(y_test, pred_svc2))
```

SVC model's precision improved from 84% to 85%

```python
from sklearn.model_selection import cross_val_score
#Now lets try to do some evaluation for random forest model using cross validation.
svc2_eval = cross_val_score(estimator = svc2, X = X_test, y = y_test, cv = 10)
svc2_eval.mean()
```
